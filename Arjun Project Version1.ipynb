### 2. About the Dataset 

The Smartphone dataset comprises 210 rows and 26 columns, detailing various attributes of different smartphone models from multiple brands.

Feature Description:

1.brand_name : Brand name of Smartphone 

2.model: Various smartphone models from different brands.

3.price: Price of diffrent smartphone models 

4.rating: Customer ratings for various models.

5.has_5g: Is the smartphone capable of connecting to a 5G network

6.has_nfc: Does the smartphone have Near Field Communication (NFC) capabilities

7.has_ir_blaster: Does the smartphone have irs blaster capabilities

8.processor_brand: Brand name of the processor 

9.num_cores: No of cores used in the processor 

10.processor_speed: Speed of the processor

11.battery_capacity: Storage Capacityof the battery

12.fast_charging_available: Whether the smartphone supporting the fast charging or not 

13.fast_charging: Smartphone charging watts

14.ram_capacity: Temproray memory storagge of the smartphone 

15.internal_memory: Smartphone storage capacity 

16.refresh_rate: Smartphone screen refresh rate 

17.num_rear_cameras: No of rear cameras in the smartphone

18.num_front_cameras: No of front cameras in the smartphone

19.os: Operatin system of the smartphone

20.primary_camera_rear: Resolution of rear primary camera

21.primary_camera_front: Resolution of front primary camera

22.extended_memory_available: whether the smartphone supporting extended memory feature or not.

23.extended_upto: Mmaximum storage capacity of extended storage.

24.resolution_width: screen resolution width. 

25.resolution_height: screen resolution height.

# <center> Univariate graphs

### <center> Model wise comparisson 

```python
plt.figure(figsize =(10,6))
sns.histplot(data = data, x = 'brand_name')
plt.xticks(rotation =75)
plt.title('Model distribution')
plt.show()
```

### <center> PRICE DISTRIBUTION

```python
sns.boxenplot(data=data, y='price')
plt.title('PRICE DISTRIBUTION')
plt.show()
```

### <center> Battery capacity Distribution

```python
sns.kdeplot(x =data.battery_capacity)
plt.title('Battery capacity Distribution')
plt.show()
```

# <center> Bivariate Graphs

### <center> Comparison between Brand and Price


```python
plt.figure(figsize =(10,8))
sns.barplot(data=data, x ='price', y ='brand_name')
plt.title('Brand VS Price')
plt.show()
```

### <center> Market treng by 5G support 

```python
sns.countplot(data=data,x ='has_5g')
plt.title('5G Trend')
plt.show()
```

### <center> Rating VS 5G Support

```python
sns.barplot(data=data, x ='has_5g', y ='rating')
plt.title('Rating VS 5G')
plt.show()
```

### <center> Processor Distribution

```python
sns.countplot(data=data, x ='processor_brand')
plt.xticks(rotation =75)
plt.title('Processor')
plt.show()
```

### <center> Rating VS Processor Speed

```python
sns.scatterplot(data=data, x ='processor_speed', y ='rating', hue ='processor_brand')
plt.title('Rating VS Processor Speed')
plt.show()
```

### <center> RAM capacity Distribution

```python
sns.countplot(data=data, x='ram_capacity')
plt.title('RAM capacity Distribution')
plt.show()
```

### <center> Rating VS Internal Memory

```python
sns.scatterplot(data=data, x ='internal_memory', y ='rating', hue ='processor_brand')
plt.title('Rating VS Internal Memory')
plt.show()
```

### <center> Rating VS Refresh Rate

```python
sns.regplot(data=data, x ='refresh_rate', y ='rating')
plt.title('Rating VS Refresh Rate')
plt.show()
```

### <center> Rating VS Rear Camera

```python
sns.regplot(data=data, x ='primary_camera_rear', y ='rating')
plt.title('Rating VS Rear Camera')
plt.show()
```

### <center> Rating VS Front Camera

```python
sns.regplot(data=data, x ='primary_camera_front', y ='rating')
plt.title('Rating VS Front Camera')
plt.show()
```

### <center> Rating VS Battery

```python
sns.lineplot(data=data, x ='rating', y ='battery_capacity')
plt.title('Rating VS Battery')
plt.show()
```

### <center> Operating  System 

```python
sns.countplot(data=data, x ='os')
plt.title('Operating System')
plt.show()
```

### <center> Screen Size VS Price

```python
sns.scatterplot(data=data,x ='price', y ='screen_size')
plt.title('ScreenSize VS Price')
plt.show()
```

## Conclusion

The analysis of the smartphone dataset has unveiled several significant trends in the smartphone market. Leading brands such as Xiaomi, Realme, Samsung, Vivo, and Oppo are dominating the market with their diverse range of models. The majority of smartphones fall within the lower to mid-range price categories. A notable finding is the positive correlation between battery capacity and price, suggesting that higher-priced smartphones generally offer superior battery performance.

Luxury smartphones are primarily produced by Huawei, Apple, and Asus. The demand for phones supporting 5G network is increasing, making them the most sought-after devices in the market. The Snapdragon processor is the most commonly used in smartphones. RAM capacities typically range from 4GB to 8GB, which are standard configurations for modern smartphones.

Furthermore, there is a positive correlation between ratings and internal memory. This dataset offers valuable insights for manufacturers, consumers, and analysts interested in the smartphone industry.
